```
x = 1 #Initialise x
lr = 0.01 #Learning rate
min = 0.000001 #minimum threshold value
delta_x = 1 #Initialising Î”x
max_iters = 1000 #Maximum number of iterations  
iters_count = 0 #Counting Iterations
gradient = []

eqn = [(1, 4), (3, 2)] #f(x) = 1*(x)**4 + 3*(x)**2 + 10

#finding gradient
for i in eqn:
  coeff, power = i[0], i[1]
  d_coeff, d_power = coeff * power, power - 1
  gradient.append((d_coeff, d_power))

#find the value of gradient at x
def gradient_at_x(value):
  global gradient
  return sum([c*(value**p) for c, p in gradient])

#finding local minima
while (abs(delta_x) > min) and (iters_count < max_iters):
  grad_value = gradient_at_x(x)
  delta_x = -lr * grad_value
  iters_count += 1
  x += delta_x
  print(f"Iteration: {iters_count}\tValue of x: {x}")

print(f"\nThe local minima occurs at: {x}")
```
