```
import math
x = 1 #Initialise x
y = 2 #Initialise y
lr = 0.01 #Learning rate
min = 0.000001 #minimum threshold value
delta_x = 1 #Initialising Î”x
delta_y = 1 #Initialising Î”y
max_iters = 1000 #Maximum number of iterations  
iters_count = 0 #Counting Iterations

#f(x) = 3x^2 + 5e^-y + 10

#find the value of gradient at x
def gradient(x_value, y_value):
  return (6 * x, -5 * math.exp(-y))

#finding local minima
while max(abs(delta_x), abs(delta_y)) > min and (iters_count < max_iters):
  grad_value_x, grad_value_y = gradient(x, y)
  delta_x = -lr * grad_value_x
  delta_y = -lr * grad_value_y
  iters_count += 1
  x += delta_x
  y += delta_y
  print(f"Iteration: {iters_count}\tValue of x: {x}, \tValue of y: {y}")

print(f"\nThe local minima occurs at: {x}, {y}")
```
